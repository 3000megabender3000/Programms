<b>Шаг 5. Проверьте гипотезы:</b>
<ol>    
<li>Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
<li>Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
<li>Задайте самостоятельно пороговое значение alpha.
<li>Поясните: Как вы сформулировали нулевую и альтернативную гипотезы;
Какой критерий применили для проверки гипотез.</ol></li>

Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Нулевая гипотеза - средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Альтернативная гипотеза: гипотеза, обратная нулевой - средние пользовательские рейтинги платформ Xbox One и PC различаются. Проверка гипотезы о равенстве среднего производим с помощью встроенных инструментов.

Для корректной проверки удалим пустые значения.

```python
df = df.dropna(subset=['user_score'])
```

```python
alpha = 0.05
sample_1 = df.query('platform=="PC"')['user_score']
sample_2 = df.query('platform=="XOne"')['user_score']
results = st.stats.ttest_ind(sample_1, sample_2, equal_var = False)
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
sample_1.mean(), sample_2.mean() # Самопроверка
```

Нулевая гипотеза не отвергнута, а это значит, что средние пользовательские рейтинги платформ Xbox One и PC почти не различаются.

Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Нулевая гипотеза - средние пользовательские рейтинги жанров Action и Sports одинаковые. Альтернативная гипотеза: гипотеза, обратная нулевой - средние пользовательские рейтинги жанров Action и Sports разные. Проверка гипотезы о равенстве среднего производим с помощью встроенных инструментов.

```python
alpha = 0.05
sample_3 = df.query('genre=="Action"')['user_score']
sample_4 = df.query('genre=="Sports"')['user_score']
results = st.stats.ttest_ind(sample_3, sample_4, equal_var = False)
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
sample_3.mean(), sample_4.mean() # Самопроверка
```

Нулевая гипотеза отвергнута, а это значит, что средние пользовательские рейтинги жанров Action и Sports различаются.
